```python
import pandas as pd
import sqlite3
```

```python
file_path='dailyActivity_merged.csv'
df=pd.read_csv(file_path)
```

```python
#create a SQLite database in memory
conn=sqlite3.connect('memory')

#load the DataFrame into the SQLite database
df.to_sql('my_table',conn,index=False,if_exists='replace')
```

```python
query="""SELECT*
FROM my_table
LIMIT 5"""
result=pd.read_sql_query(query,conn)
result
```

```python
query="""SELECT COUNT(Id)
FROM my_table
GROUP BY Id"""
result=pd.read_sql_query(query,conn)
result
```

```python
query="""
SELECT Id,COUNT(Id) AS total_logged_uses, CASE
WHEN COUNT(Id) BETWEEN 25 AND 31 THEN 'active_user'
WHEN COUNT(Id) BETWEEN 15 AND 24 THEN 'moderate_user'
WHEN COUNT(Id) BETWEEN 0 AND 14 THEN 'light_user'
END usage_time 
FROM my_table
GROUP BY Id"""
result=pd.read_sql_query(query,conn)
result
```

```python
query="""SELECT Id,
MIN(TotalSteps) AS min_total_steps,
MAX(TotalSteps) AS max_total_steps,
AVG(TotalSteps) AS avg_total_steps,
MIN(Calories) AS min_calories,
MAX(Calories) AS max_calories,
AVG(Calories) AS avg_calories
FROM my_table
GROUP BY Id"""
result=pd.read_sql_query(query,conn)
result
```

```python
query="""SELECT Id,
AVG(VeryActiveMinutes) AS Avg_very_act,
AVG(FairlyActiveMinutes) AS avg_fairly_act,
AVG(LightlyActiveMinutes) AS avg_lightly_act
FROM my_table
GROUP BY Id"""
result=pd.read_sql_query(query,conn)
result
```

```python
query="""SELECT Id,
AVG(VeryActiveMinutes)+AVG(FairlyActiveMinutes)+AVG(LightlyActiveMinutes) AS sum_avg,
CASE 
WHEN AVG(VeryActiveMinutes)+AVG(FairlyActiveMinutes)+AVG(LightlyActiveMinutes) >=150 THEN 'meet'
WHEN AVG(VeryActiveMinutes)+AVG(FairlyActiveMinutes)+AVG(LightlyActiveMinutes) <150 THEN 'Dont meet'
END is_meet
FROM my_table
GROUP BY Id"""
result=pd.read_sql_query(query,conn)
result
```

```python
query="""SELECT Id,
SUM(VeryActiveMinutes+FairlyActiveMinutes+LightlyActiveMinutes) AS active_minutes,
CASE 
WHEN SUM(VeryActiveMinutes+FairlyActiveMinutes+LightlyActiveMinutes) >150 THEN 'meet'
ELSE 'doesnt meet'
END recommendation
FROM my_table
WHERE ActivityDate BETWEEN '5/1/2016' AND '7/1/2016'
GROUP BY Id
"""
result=pd.read_sql_query(query,conn)
result
```
